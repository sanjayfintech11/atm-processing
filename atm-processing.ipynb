# atm-processing

Use the "Run" button to execute the code.

```python
# define function to show menu and get user input
def show_menu():
    print("Welcome to the ATM!")
    print("Please choose an option:")
    print("1. Show balance")
    print("2. Deposit")
    print("3. Withdraw")
    print("4. Exit")
    return input("Option: ")



# initialize account balance and withdrawal limit
balance = 1000
LIMIT = 500
MIN_BALANCE = 100


# define function to get user PIN
def get_pin():
    return input("Enter your PIN: ")

# define function to handle deposit
def deposit(amount):
    global balance
    balance += amount
    print("Deposit successful. Your new balance is ", balance)

# define function to handle withdrawal
def withdraw(amount):
    global balance
    if amount > LIMIT:
        print("Withdrawal limit exceeded")
    elif balance - amount < MIN_BALANCE:
        print("Minimum balance must be maintained")
    else:
        balance -= amount
        print("Withdrawal successful. Your new balance is ", balance)

# main program
pin = get_pin()
if pin == "1234":
    while True:
        option = show_menu()
        if option == "1":
            print("Your balance is ", balance)
        elif option == "2":
            amount = int(input("Enter deposit amount: "))
            deposit(amount)
        elif option == "3":
            amount = int(input("Enter withdrawal amount: "))
            withdraw(amount)
        elif option == "4":
            print("Goodbye!")
            break
        else:
            print("Invalid option. Please try again.")
else:
    print("Invalid PIN. Please try again.")
```
